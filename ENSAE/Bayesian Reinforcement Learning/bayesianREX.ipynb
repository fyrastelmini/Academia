### Generating the posterior Reward function and training the ai using it (this should take a few minutes)

This script reads the demonstration data provided in the previous step, and uses it to infer and sample from the posteriori reward distribution.

```python
!python LinearFeatureMCMC_auxiliary.py --env_name hero --models_dir ./models/ --weight_outputfile ./mcmc_data/hero_mcmc.txt --num_mcmc_steps 200000 --map_reward_model_path ./mcmc_data/hero_map.params --pretrained_network ./pretrained_networks/hero_pretrained.params_stripped.params --encoding_dims 64
```

We run the reinforcement learning training with the reward function generated above (this doesnt take too long since we're not going to train it all the way, this is just to demonstrate the behaviour of the agent right at the start of the training using the bayesian reward funtion, which should be bad)

```python
!OPENAI_LOG_FORMAT='stdout,log,csv,tensorboard' OPENAI_LOGDIR=~/tflogs/hero_mean python -m baselines.run --alg=ppo2 --env=HeroNoFrameskip-v4 --custom_reward mcmc_mean --custom_reward_path ../mcmc_data/hero_map.params --mcmc_chain_path ../mcmc_data/hero_mcmc.txt --seed 0 --num_timesteps=1e4  --save_interval=50 --num_env 9 --embedding_dim 64
```
